```python
import pandas as pd
df = pd.read_csv("/content/HR-Employee-Attrition.csv")
df
```

```python
### Cleaning steps
# check for missing values
print("\nmissing values:")
print(df.isnull().sum())

# Removing duplicate:
df.dropna(inplace=True)
print("\nThere is no duplicate values")

# Handle missing values
df.fillna(df.mean(numeric_only=True), inplace=True)
print("\nThere is no missing values")
```

```python
### Statistical Analysis:
# Group by and aggregate data.
print("\naverage age by gender:")
print(df.groupby('Gender')['Age'].mean().reset_index())
print("\naverage years spents at company by gender:")
print(df.groupby('Gender')['YearsAtCompany'].mean().reset_index())
print("\n Attrition count by departments:")
print(df.groupby('Department')['Attrition'].value_counts())
print("\nmultiple aggregate functions:")
print(df.groupby('Department')['MonthlyIncome'].agg(["mean","max","min"]).reset_index())


# Compute basic statistics: mean, median, mode, std, variance.
df.describe()
print("Mean Age:", df['Age'].mean())
print("Median Age:", df['Age'].median())
print("Mode Age:", df['Age'].mode()[0])
print("Standard Deviation (MonthlyIncome):", df['MonthlyIncome'].std())
print("Variance (MonthlyIncome):", df['MonthlyIncome'].var())
print("Median Salary:", df['MonthlyIncome'].median())

''' Average age of female employees is 37Years
Average age of male employees is 36 years.'''

''' Median Monthly Imcome is 6,000'''

'''Income distribution shows moderate spread (std ≈ 4707)'''
```

```python
### Visualizations:
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Line chart:

AverageMonthlyIncome = df.groupby('Department')['MonthlyIncome'].mean().reset_index()
plt.figure(figsize = (8,4))
plt.plot(AverageMonthlyIncome['Department'], AverageMonthlyIncome['MonthlyIncome'],marker='o', linestyle='-.',color='blue')

plt.title("Average monthly income by department")
plt.xlabel('Department')
plt.ylabel('AverageMonthlyIncome')
plt.tight_layout()
plt.show()
```

```python
# plt.figure(figsize=(6,4))

plt.scatter(df.Age, df.MonthlyIncome, color='green', edgecolor='black',marker='x',alpha=1.0)

plt.title("Monthly income by age")
plt.xlabel('Age')
plt.ylabel('MonthlyIncome')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Pie chart:
employeecount = df["Department"].value_counts().reset_index()
employeecount.columns = ['Department','Count']
print(employeecount)
print(employeecount.columns)

plt.figure(figsize=(6,4))
plt.pie(employeecount['Count'],
        labels=employeecount['Department'],
        colors=['Pink','lightgreen','lightblue'],
        autopct="%1.2f%%",
        explode=[0.08,0,0],
        shadow=True
        )
plt.title("Department wise Employee count ")
plt.show()
```

```python
# Histogram Chart:
plt.figure(figsize=(8,5))
plt.hist(df['Age'],bins=15, color='lightgreen',edgecolor='black')
plt.grid(axis='y',linestyle='-.')
plt.title("Distribution of Employee age")
plt.xlabel('Age')
plt.ylabel("Number of Employees")
plt.tight_layout()
plt.show()
```

```python
# bar chart:
import numpy as np
attrition_counts = df.groupby(['Department', 'Attrition']).size().unstack(fill_value=0)
plt.figure(figsize=(8,4))

x = np.arange(len(attrition_counts.index))
width = 0.35

plt.bar(x - width/2, attrition_counts['Yes'], width,
        color='lightgreen',label='Yes', edgecolor='black', linewidth=1.8)
plt.bar(x + width/2, attrition_counts['No'], width,
        color='lightblue',label='No', edgecolor='black', linewidth=1.8)


plt.title("Attrition Count by Department")
plt.xlabel("Department")
plt.ylabel("Number of Employees")
plt.tight_layout()
plt.show()
```

```python
# box plot:
plt.figure(figsize=(6,5))
plt.boxplot(df['Age'], patch_artist=True,boxprops=dict(facecolor='pink', color='black'),
            medianprops=dict(color='red', linewidth=2))
plt.title("Box Plot of Monthly Income", fontsize=14)
plt.ylabel("Monthly Income")
plt.tight_layout()
plt.show()
```

```python
# Webscrabbing:
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd

url = "https://en.wikipedia.org/wiki/List_of_largest_companies_in_India"

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
                  "AppleWebKit/537.36 (KHTML, like Gecko) "
                  "Chrome/120.0.0.0 Safari/537.36"
}

response=requests.get(url, headers=headers)
soup = bs(response.text, "html.parser")

table =soup.find('table')

headers = table.find('tr').text.split('\n')

headings=[]
for i in range(len(headers)):
  if i%2==1:
    headings.append(headers[i])

display(headings)


trs = table.find_all('tr')[1:]

rows=[]

for tr in trs:
  row=[]
  for td in tr.find_all(['td','th']):
    text=td.get_text(strip=True)
    row.append(text)
  rows.append(row)

rows

for r  in rows:
  print(r)

df = pd.DataFrame(rows, columns=headings)

df_main=pd.read_csv("/content/HR-Employee-Attrition.csv")
combined_df = pd.concat([df_main,df], axis=1)
display(combined_df)

```
